```python
import pandas as pd
import re
```

```python
file1 = pd.read_csv("Accelerating_Star_Ages_Yasmin.csv")
file2 = pd.read_csv("Filtered_Accelerating_Stars_Data.csv")

col1 = "In TESS"  
col2 = "TICID"      
```

```python
def clean_tic(value):
    if pd.isna(value):
        return None
    
    value = str(value).lower().replace("tic", "")
    digits = re.sub(r"\D", "", value)  
    return digits
```

```python
tic_ids_1 = set(file1[col1].apply(clean_tic).dropna())
tic_ids_2 = set(file2[col2].apply(clean_tic).dropna())
```

```python
common_tic_ids = sorted(tic_ids_1.intersection(tic_ids_2))
```

```python
print(f"Number of common TIC IDs: {len(common_tic_ids)}")
print(common_tic_ids)
```

```python
pd.DataFrame(common_tic_ids, columns=["tic_id"]).to_csv("common_tic_ids.csv", index=False)
```
